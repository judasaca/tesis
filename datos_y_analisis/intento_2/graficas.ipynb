```python
import numpy as np 
import matplotlib.pylab as plt
import pandas as pd
plt.rcParams.update({"font.size":18})
```

```python
def fourier(tiempo,voltaje):
    t_max=np.max(tiempo)
    Ts=t_max/len(tiempo)
    Fs=1/Ts
    amplitude=np.abs(np.fft.fft(voltaje))/(len(tiempo)/2)
    F=Fs*np.arange(0,len(voltaje))/len(voltaje)
    return F[0:len(F)//2], amplitude[0:len(F)//2]
```

```python
points=10000
v=np.zeros(points)
i=1
while i<points:
    n=np.random.choice([-1,1])
    v[i]=v[i-1]+n
    i+=1
```

```python
t=np.arange(0,len(v))
f, a = fourier(t,v)
psd= a**2/points
```

```python
plt.figure(figsize=(14,8))
plt.plot(t,v)
plt.xlabel("t (s)")
plt.ylabel("v")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.loglog(f,a)
plt.xlabel("f(Hz)")
plt.ylabel(r"DEP ($V^2/Hz$)")
plt.grid(True)
plt.show()
```
