```python
import numpy as np 
import matplotlib.pylab as plt 
import pandas as pd
from scipy.stats import kde
import seaborn as sns
sns.set_style("darkgrid")
plt.rcParams.update({"font.size":18})
```

```python
carpeta="datos_pseudotermica_deteccion_directa/"
n=["2V", "3V", "4V", "5V","6V","7V", "8V","9V", "10V"]
plt.figure(figsize=(18,10))
for i in range(len(n)):
    datos=pd.read_csv(carpeta+n[i]+".csv")
    #plt.hist(datos["CH1"], bins=10, label=n[i], alpha=1-0.05*i, density=True)
    density = kde.gaussian_kde(datos["CH1"])
    x=np.linspace(min(datos["CH1"]), max(datos["CH1"]), 1000)
    plt.plot(x,density(x), label=n[i], linewidth=3)


plt.xlabel("Voltaje de salida del diodo", fontsize=18)
plt.xlim(0,0.3)
plt.title("Histogramas de fuente pseudotérmica con el disco a distintos voltajes", fontsize=18)
plt.legend(fontsize=18)
```

```python
carpeta="datos_pseudotermica_deteccion_directa/"
plt.figure(figsize=(18,10))
x=[]
y=[]
for i in range(len(n)):
    datos=pd.read_csv(carpeta+n[i]+".csv")
    x.append(int(n[i][:-1]))
    y.append(datos["CH1"].std())
plt.scatter(x,y, s=90)
#plt.xlim(0,0.3)
plt.legend(fontsize=18)
plt.xlabel("Voltaje del disco de acrilico",fontsize=20)
plt.ylabel(r'$\sigma$ del voltaje de salida del fotodiodo', fontsize=20)
```

```python
carpeta="datos_pseudotermica_deteccion_directa/"
plt.figure(figsize=(18,10))
x=[]
y=[]
for i in range(len(n)):
    datos=pd.read_csv(carpeta+n[i]+".csv")
    x.append(int(n[i][:-1]))
    y.append(datos["CH1"].mean())
plt.scatter(x,y)
#plt.xlim(0,0.3)
plt.legend(fontsize=18)
```
