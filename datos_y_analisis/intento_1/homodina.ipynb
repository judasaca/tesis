```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import pandas as pd
sns.set_style("darkgrid")
plt.rcParams.update({"font.size":18,
            "figure.figsize":[17,7]})
```

```python
def load_data(name):
    p="datos_homodina"
    return pd.read_csv("{}/{}.csv".format(p,name),skiprows=14)
datos=load_data("encendido")
homodino=load_data("homodino")
len(datos)
```

```python
def load_data2(name, p="spectrum_analizer"):
    return pd.read_csv("{}/{}.csv".format(p,name),skiprows=9, usecols=["TraceA(Hz)", "TraceA(dBm)"])
dark_noise=load_data("dark_noise")
```

```python
plt.figure()
sns.histplot(datos["CH1"],bins=30, kde=True, 
             label=r'Local oscilator $\sigma={}$'.format(round(datos["CH1"].std(),4)),
            alpha=1,
            color="blue",
            stat="density")
sns.histplot(dark_noise["CH1"], bins=30, alpha=0.5, 
             label='Dark noise $\sigma={}$'.format(round(dark_noise["CH1"].std(),4)), 
             kde=True,
            color="red",
            stat="density")
sns.histplot(homodino["CH1"], bins=30, alpha=0.4, 
             label='Signal y local oscilator $\sigma={}$'.format(round(homodino["CH1"].std(),4)), 
             kde=True,
            color="green",
            stat="density")
plt.xlabel("Voltaje (V)")
plt.title("Histograma de voltajes")
plt.ylabel("")
plt.legend()
plt.show()
```
