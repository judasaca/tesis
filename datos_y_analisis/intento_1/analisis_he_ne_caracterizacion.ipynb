```python
import matplotlib.pylab as plt
import numpy as np 
import pandas as pd
import seaborn as sns 
sns.set_style("darkgrid")
plt.rcParams.update({'font.size': 18})
```

```python
path="datos_he_ne_deteccion_directa_caracteriacion_fotodiodo"
resistencias=[0.894, 0.750, 0.660, 0.550, 0.440, 0.362, 0.230, 0.144, 0.043]
```

```python
promedios=[]
desviaciones=[]
for i in range(1,10):
    filepath=path+"/"+str(i)+".csv"
    datos=pd.read_csv(filepath)
    promedio=datos["CH1"].mean()
    desviacion=datos["CH1"].std()
    promedios.append(promedio)
    desviaciones.append(desviacion)    
```

```python
plt.figure(figsize=(20,15))

plt.subplot(2,1,1)
plt.title("Promedios")
plt.scatter(resistencias, promedios, s=80)
plt.plot(resistencias, promedios)
plt.xlabel(r'Resistencia (M$\Omega$)')
plt.ylabel(r'$<V>$ (V)')

plt.subplot(2,1,2)
plt.title("Desviacin estandar")
plt.scatter(resistencias, desviaciones,s=80)
plt.plot(resistencias, desviaciones)
plt.xlabel(r'Resistencia (M$\Omega$)')
plt.ylabel(r'$\sigma$')

```
