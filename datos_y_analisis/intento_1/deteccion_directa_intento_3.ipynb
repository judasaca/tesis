```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import pandas as pd
sns.set_style("darkgrid")
plt.rcParams.update({"font.size":18,
            "figure.figsize":[17,7]})
```

```python
def load_data(name):
    p="datos_deteccion_directa_intento_3"
    return pd.read_csv("{}/{}.csv".format(p,name),skiprows=14)
encendido=load_data("laser_encendido")
dark_noise=load_data("dark_noise")
pseudotermica=load_data("pseudotermica")
```

```python
plt.figure()
sns.histplot(encendido["CH1"],bins=30, kde=True, 
             label=r'Laser He-Ne $\sigma={}$'.format(round(encendido["CH1"].std(),4)),
            alpha=1,
            color="blue",
            stat="density")
sns.histplot(dark_noise["CH1"], bins=30, alpha=0.5, 
             label='Dark noise $\sigma={}$'.format(round(dark_noise["CH1"].std(),4)), 
             kde=True,
            color="red",
            stat="density")
sns.histplot(pseudotermica["CH1"], bins=30, alpha=0.3, 
             label='Pseudo-térmica $\sigma={}$'.format(round(pseudotermica["CH1"].std(),4)), 
             kde=True,
            color="green",
            stat="density")

plt.xlabel("Voltaje (V)")
plt.ylabel("")
plt.title("Comparación de ruido de detección directa en un fotodiodo de caja metálica a máxima ganancia")
plt.legend()
plt.show()
```
