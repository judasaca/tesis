```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
import seaborn as sns
from scipy.signal import filtfilt
import scipy as sp
sns.set_style("darkgrid")
plt.rcParams.update({"font.size":18,
                    "figure.figsize":[17,8]})
```

```python
w=np.linspace(0,5e4)
R=1000
C=0.1e-6

u=w*R*C
H=np.sqrt(u**4+u**2)/(1+u**2)
wc=10e3
uc=wc*R*C
Hc=np.sqrt(uc**4+uc**2)/(1+uc**2)
```

```python
plt.figure()

labels=["0","10", "20", "30", "40", "50"]
values=[0,10e3,20e3,30e3,40e3,50e3]
plt.plot(w,H)
plt.plot([10e3,10e3,-2.07e3], [-0.1,Hc,Hc], "--", label=r'$\omega_c$')
plt.xticks(values,labels)
plt.xlabel("w (KHz)")
plt.ylabel(r'$|H(\omega)|$')
plt.ylim(-0.07,1.07)
plt.xlim(-1e3,41e3)
plt.legend()
plt.show()
```

# Filtros digitales

```python
def pasabanda(señal):
    
    corte_bajo=100
    corte_alto=150
    
    delta_t=1/len(señal)
    
    nyq= 0.5/delta_t
    bajo=corte_bajo/nyq
    alto=corte_alto/nyq
    
    grado=2
    b,a = sp.signal.butter(grado,[bajo,alto], "bandpass", analog=False)
    y=filtfilt(b,a, señal)
    
    w, h = sp.signal.freqs(b, a)
    plt.semilogx(w, 20 * np.log10(abs(h)))
    plt.title('Butterworth filter frequency response')
    plt.xlabel('Frequency [radians / second]')
    plt.ylabel('Amplitude [dB]')
    plt.margins(0, 0.1)
    plt.grid(which='both', axis='both')
    plt.axvline(100, color='green') # cutoff frequency
    plt.show()
    return y
    
```

```python
def fourier(tiempo,voltaje):
    t_max=np.max(tiempo)
    Ts=t_max/len(tiempo)
    Fs=1/Ts
    amplitude=np.abs(np.fft.fft(voltaje))/(len(tiempo)/2)
    F=Fs*np.arange(0,len(voltaje))/len(voltaje)
    return F[0:len(F)//2], amplitude[0:len(F)//2]
```

```python
w_principal=[120,130]
A_principal=[10,15]

n_noise=20
A_max_noise=15
n_points=4000
A_noise=np.random.random(n_noise)*10
w_noise=np.concatenate((np.random.random(int(n_noise/2))*50,np.random.random(int(n_noise/2))*50+200))
t=np.linspace(0,1,n_points)

y=np.zeros(len(t))
for i in range(len(w_principal)):
    y=y+A_principal[i]*np.cos(2*np.pi*w_principal[i]*t)
for i in range(len(A_noise)):
    y=y+A_noise[i]*np.cos(2*np.pi*w_noise[i]*t)
y=y+np.random.random(n_points)*30-15
```

```python
y_filtrado=pasabanda(y)
```

```python
plt.figure(figsize=(17,13))

plt.subplot(2,1,1)
plt.plot(t,y, label="Señal sin filtrar")
plt.plot(t,y_filtrado, label="Señal filtrada", alpha=0.9)
plt.legend()
plt.xlabel("t")
plt.ylabel("Amplitud")

f_filtrada, a_filtrada=fourier(t, y_filtrado)
f_sin_filtrar, a_sin_filtrar=fourier(t, y)
plt.subplot(2,1,2)
plt.plot(f_sin_filtrar, a_sin_filtrar, label="Espectro de la señal sin filrar")
plt.plot(f_filtrada, a_filtrada, label="Espectro de la señal filtrada")
plt.xlim(-10,300)
plt.xlabel("f(Hz)")
plt.ylabel("Amplitud")
plt.legend()
plt.show()
```
