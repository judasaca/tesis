```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
import seaborn as sns
sns.set_style("darkgrid")
plt.rcParams.update({"font.size":18,
                    "figure.figsize":[17,8]})
```

```python
w=np.linspace(0,5e4)
R=1000
C=0.1e-6

u=w*R*C
H=np.sqrt(u**4+u**2)/(1+u**2)
wc=10e3
uc=wc*R*C
Hc=np.sqrt(uc**4+uc**2)/(1+uc**2)
```

```python
plt.figure()

labels=["0","10", "20", "30", "40", "50"]
values=[0,10e3,20e3,30e3,40e3,50e3]
plt.plot(w,H)
plt.plot([10e3,10e3,-2.07e3], [-0.1,Hc,Hc], "--", label=r'$\omega_c$')
plt.xticks(values,labels)
plt.xlabel("w (KHz)")
plt.ylabel(r'$|H(\omega)|$')
plt.ylim(-0.07,1.07)
plt.xlim(-1e3,41e3)
plt.legend()
plt.show()
```
