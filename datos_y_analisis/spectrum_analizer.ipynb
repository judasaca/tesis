```python
import numpy as np 
import matplotlib.pylab as plt
import pandas as pd
import seaborn as sns
plt.rcParams.update({'font.size':18,
             'figure.figsize':[17,8]})
sns.set_style(style='darkgrid')
```

# Ruido oscuro

```python
def load_data(name):
    p="spectrum_analizer"
    return pd.read_csv("{}/{}.csv".format(p,name),skiprows=9, usecols=["TraceA(Hz)", "TraceA(dBm)"])
data = pd.read_csv("spectrum_analizer/dark_noise.csv",skiprows=9, usecols=["TraceA(Hz)", "TraceA(dBm)"])
frecuency=data["TraceA(Hz)"]
power=data["TraceA(dBm)"]
```

```python
plt.plot(frecuency, power)
plt.xlabel("Hz")
plt.ylabel("dBm")
plt.show()
```

```python
sns.histplot(power[power<-70], bins=50)
plt.show()
```

# Respuesta en frecuencia del laser He-Ne encendido y apagado.

```python
on=load_data("deteccion_directa_laser_on")
off=load_data("deteccion_directa_laser_off")
```

```python
fig=on.plot(x="TraceA(Hz)", y="TraceA(dBm)", label="Laser on")
fig.plot(off["TraceA(Hz)"], off["TraceA(dBm)"], label="Laser off")
plt.legend()
plt.show()
```

# Comparacion de fotodiodos

```python
sin_caja_metalica=load_data("comparacion_fotodiodos_sin_caja_metalica")
con_caja_sin_filtro=load_data("comparacion_fotodiodos_con_caja_sin_filtro")
con_caja_con_filtro=load_data("comparacion_fotodiodos_con_caja_con_filtro")
```

```python
plt.plot(sin_caja_metalica["TraceA(Hz)"],sin_caja_metalica["TraceA(dBm)"], label="Caja plástica")
plt.plot(con_caja_sin_filtro["TraceA(Hz)"],con_caja_sin_filtro["TraceA(dBm)"], label="Caja metálica sin filtro")
plt.plot(con_caja_con_filtro["TraceA(Hz)"],con_caja_con_filtro["TraceA(dBm)"], label="Caja metálica con filtro")
plt.xlabel("Hz")
plt.ylabel("dBm")
plt.legend()
plt.show()
```
